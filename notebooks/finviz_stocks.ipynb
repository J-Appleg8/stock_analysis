```python
import requests
import pandas as pd
import json
from bs4 import BeautifulSoup
from tqdm.auto import tqdm

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
          'Upgrade-Insecure-Requests': '1', 'Cookie': 'v2=1495343816.182.19.234.142', 'Accept-Encoding': 'gzip, deflate, sdch',
           'Referer': "http://finviz.com/quote.ashx?t="}

def get_fundamental_data(df, symbol:str):
    try:      
        r = requests.get("http://finviz.com/quote.ashx?t=" + symbol.lower(), headers=headers)
        soup = BeautifulSoup(r.content, 'html.parser') 
        for m in df.columns:                
            df.loc[symbol, m] = soup.find(string=m).find_next(class_='snapshot-td2').text                
    except Exception as e:
        print (symbol, 'not found')
        print(e)

    return df


def get_financial_data(symbol: str, statement:str):
    try:
        response = requests.get(f"https://elite.finviz.com/api/statement.ashx?t={symbol}&s={statement}", headers=headers).text
        response_dict = json.loads(response)['data']
        df = pd.DataFrame(response_dict).transpose()
        df.columns = df.iloc[0]
        df.drop(['Period Length', 'Period End Date'], inplace=True, errors='ignore')

        cols = [i for i in df.columns if i not in ['Period End Date']]
        for col in cols:
            df[col] = pd.to_numeric(df[col].str.replace(',', '')).round(2).fillna(0)
    except Exception as e:
        print (symbol, 'not found')
        print(e)

    return df


metric = [
    'Price',
	'P/B',
	'P/E',
	'Forward P/E',
	'PEG',
	'Debt/Eq',
	'EPS (ttm)',
	'EPS next 5Y',
	'Dividend %',
	'ROE',
	'ROI',
	'EPS Q/Q',
	'Insider Own'
]

fundamentals_df = pd.DataFrame(columns=metric)
fundamentals_df = get_fundamental_data(fundamentals_df, 'ASRT')
fundamentals_df = fundamentals_df.replace('-', 0)

income_statement_df = get_financial_data('ASRT', 'IA').copy()
balance_sheet_df = get_financial_data('ASRT', 'BA').copy()
cash_flow_df = get_financial_data('ASRT', 'CA').copy()

income_statement_df.loc['Test'] = income_statement_df.apply(lambda x: x['Cost of Revenue'] + x['Gross Profit'])

fundamentals_df
income_statement_df
```
